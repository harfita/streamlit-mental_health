```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

#untuk mengabaikan warning
warnings.filterwarnings("ignore")

print("Semua Library diatas termuat")
```

```python
df = pd.read_csv("mental_health_diagnosis_treatment_.csv")
df
```

```python
df_baru = df.copy()
df_baru = df_baru[[
    "Age",
    "Symptom Severity (1-10)",
    "Mood Score (1-10)",
    "Sleep Quality (1-10)",
    "Physical Activity (hrs/week)",
    "Adherence to Treatment (%)",
    "Stress Level (1-10)",
    "Gender",
    "Diagnosis",
    "Medication",
    "Therapy Type",
    "AI-Detected Emotional State",
    "Treatment Progress (1-10)"
    
]]

df_baru
```

```python
from sklearn.preprocessing import LabelEncoder

df_baru['Gender'] = LabelEncoder().fit_transform(df['Gender'])
df_baru['Diagnosis'] = LabelEncoder().fit_transform(df['Diagnosis'])
df_baru['Medication'] = LabelEncoder().fit_transform(df['Medication'])
df_baru['Therapy Type'] = LabelEncoder().fit_transform(df['Therapy Type'])
df_baru['AI-Detected Emotional State'] = LabelEncoder().fit_transform(df['AI-Detected Emotional State'])
df_baru['Outcome'] = LabelEncoder().fit_transform(df['Outcome'])
```

```python
# Melatih model
from sklearn.model_selection import train_test_split

#memasukkan linear regression
from sklearn.linear_model import LinearRegression
#import GBR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor

#metrics evaluasi
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#membagi X dan Y
X = df_baru.drop(columns=["Treatment Progress (1-10)"])
y = df_baru["Treatment Progress (1-10)"]
```

```python
 # Setting rasio split (40% data untuk pengujian)
test_size_ratio = 0.4

# Membagi dataset menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size_ratio, random_state=42)

# Memuat model Random Forest Regressor
model_rfr = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model
model_rfr.fit(X_train, y_train)

# Memprediksi nilai untuk data training dan testing
y_pred_train = model_rfr.predict(X_train)
y_pred_test = model_rfr.predict(X_test)

# Menghitung metrik evaluasi
rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

# Menampilkan hasil evaluasi model
print(f"R2_score pelatihan RFR: {r2_score(y_train, y_pred_train)}")
print(f"R2_score pengujian RFR: {r2_score(y_test, y_pred_test)}")
print(f"MSE pelatihan RFR: {mean_squared_error(y_train, y_pred_train)}")
print(f"MSE pengujian RFR: {mean_squared_error(y_test, y_pred_test)}")
print(f"RMSE pengujian RFR: {rmse}")
```
